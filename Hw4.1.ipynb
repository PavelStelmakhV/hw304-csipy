```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy import integrate
```

```python
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
```

```python
time = np.linspace(0, 11, 12)
time
```

```python
plt.scatter(x=time, y=speed)
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.grid()
```

## kind='cubic'

```python
f_cubic = interp1d(time, speed, kind='cubic')
plt.plot(time, speed, 'o')
more_date = np.linspace(0, 11, 10000)
plt.plot(more_date, f_cubic(more_date))
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.grid()
```

```python
result = integrate.quad(f_cubic, 0, 11)
result[0]
```

## kind='quadratic'

```python
f_quadratic = interp1d(time, speed, kind='quadratic')
plt.plot(time, speed, 'o')
more_date = np.linspace(0, 11, 10000)
plt.plot(more_date, f_quadratic(more_date))
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.grid()
```

```python
result = integrate.quad(f_quadratic, 0, 11)
result[0]
```
